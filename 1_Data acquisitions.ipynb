Since some of the data files acquired are too large to import in excel/google sheets, I used Python and pandas to select smaller amount of data that we're interested in. 

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///data/restaurants.db")
```

```python
df = pd.read_sql("""SELECT * FROM calls_data WHERE "Complaint Type"="Rodent";""", engine, parse_dates=["Created Date"])
df.head()
```

```python
calls_df = df[df['Created Date']> "1/1/2019"]
calls_df.shape
```

```python
calls_df.to_csv("./data/rodent_calls.csv", index=False)
```

```python
df2 = pd.read_sql("""SELECT * FROM Inspection_data WHERE "VIOLATION CODE"="04K";""", engine, parse_dates=["INSPECTION DATE"])
df2.head()
```

```python
df2.to_csv("./data/inspections.csv", index=False)
```
